```python
import pandas as pd 
import matplotlib.pyplot as plt

```

```python
# US Weather records over a 12 month period, average temperatures taken from recoreds since 1880
# Dataset from https://fivethirtyeight.com/features/what-12-months-of-record-setting-temperatures-looks-like-across-the-u-s/


data = pd.read_csv("weather.csv") 
data['date'] = data['date'].apply(pd.to_datetime)
```

```python
# Calculate mean and median values for the actual precipitation, max and min temperatures

print("Mean of actual precipitation ",data["actual_precipitation"].mean())
print("Median of actual precipitation ",data["actual_precipitation"].median())
print("Maximum actual precipitation", data["actual_precipitation"].max())
print("Minimum actual precipitation", data["actual_precipitation"].min())

print(" \nMean of min temp ",data["actual_min_temp"].mean())
print("Median of min temp ",data["actual_min_temp"].median())

print(" \nMean of max temp ",data["actual_max_temp"].mean())
print("Median of max temp ",data["actual_max_temp"].median())

```

```python
#Box plot for actual precipitation showing outliers
import matplotlib.pyplot as plt
fig1, ax1 = plt.subplots()
ax1.set_title('Actual Precipitation')
ax1.boxplot(data["actual_precipitation"])
plt.ioff()
plt.show()
```

```python
#Actual precipitation box plot without outliers
fig1, ax1 = plt.subplots()
ax1.set_title('Actual Precipitation')
ax1.boxplot(data["actual_precipitation"], showfliers=False)
plt.ioff()
plt.show()
```

```python
# Actual precipitation bar chart
plt.bar(data["date"],data["actual_precipitation"], color='green')
plt.xlabel("Actual Precipitation")
plt.ylabel("Precipitation (in)")
plt.title("Date")
```

```python
temps=[data["actual_min_temp"], data["actual_max_temp"]] 
fig3, ax2 = plt.subplots() 
ax2.set_title('Max and Min Temperatures') 
ax2.boxplot(temps, notch=True)

plt.show()

```

```python
#Line graph to show max and min temperatures
plt.plot(data['date'], data['actual_min_temp'] ,  'blue')
plt.plot(data['date'], data['actual_max_temp'] ,  'red')
label="Max and Min Temperatures (F)"
plt.legend()
```
